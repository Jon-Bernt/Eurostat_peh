```python
import pandas as pd
import numpy as np
```

# Importing dataset from Eurostat from bulk download facility

```python
url="https://ec.europa.eu/eurostat/estat-navtree-portlet-prod/BulkDownloadListing?sort=1&file=data%2Fnrg_bal_peh.tsv.gz"
url_data=pd.read_csv(url, sep="\t")
url_data.head()
```

#Split column "unit,product,indic_nrg,geo\time" by comma

```python
url_data_split = url_data[url_data.columns[0]].str.split(",", expand = True)
url_data_split.head()
url_data_split.shape
```

#Add to original dataframe

```python
new_url_data= pd.concat([url_data_split, url_data.iloc[:, 1:]], axis=1)
new_url_data.tail()
```

#Transpose years and values (converting to columns)

```python
new_url_data = pd.melt(new_url_data,id_vars=[0,1,2,3])
new_url_data.shape
```

```python
#Split column "variable" into year and month
```

```python
new_url_data = new_url_data[["variable", 2,0,1,3, "value"]]
new_url_data_cols = ["Year", "Unit", "nrg_bal", "siec", "Country", "Value"]
new_url_data.columns = new_url_data_cols
new_url_data.head()
```

# Import Eurostat country code list from local directory

# Merging Eurostat dataset with Eurostat product list and country codes

```python
#Merge dataset with product codes and description

df_merge = pd.merge(new_url_data, df_description_list, on = "siec")
df_merge.head()
```

```python
df_merge.shape #same shape as raw data file
```

```python
#Merge with country codes and countries
df_merge = pd.merge(df_merge, df_country_codes, left_on = "Country", right_on = "Country_code")
df_merge.head()
```

```python
df_merge.shape
```

```python
#Drop column "Country_x"
df_merge.drop("Country_x", axis = 1, inplace = True)
```

```python
df_merge.rename({"Country_y" : "Country"}, axis = 1, inplace = True)
df_merge.rename({"nrg_bal_y" : "nrg_Description"}, axis = 1, inplace = True)
df_merge.rename({"nrg_bal_x" : "nrg_bal"}, axis = 1, inplace = True)
df_merge.head()
```

```python
#Changing the order of columns
df_merge = df_merge[["siec", "nrg_bal", "nrg_Description", "Country_code", "Country", "Year", "Description","Unit", "Value"]]
df_merge.head()
```

```python
#Convert "Value" from object type to numeric
df_merge.Value = pd.to_numeric(df_merge.Value, errors = "coerce")
df_merge.head()
```

```python
df_merge.dtypes
```

```python
#Filter on "nrg_bal" == "GEP" (Gross electricity production)
df_merge = df_merge.loc[df_merge["nrg_bal"] == "GEP"]
```
